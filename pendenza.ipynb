# Definizione della funziona $f(x)$

$f(x) = x^3 + x^2 - 5x + 2$

```python
def f(x):
    return(x**3 + x**2 - 5*x + 2)
```

```python
x = 0
f(x)
```

Genero il vettore degli argomenti $x$

```python
x = np.linspace(-10 , 10 , 201)
print(x)
```

Genero il vettore delle immagini $y$

```python
y = f(x)
print(y)
```

```python
# Traccio il grafico di f(x) limitato da -10 a 10 sia in x che in y, senza punti
import matplotlib.pyplot as plt
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Grafico di f(x) = x^3 + x^2 - 5x + 2')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid(True)
plt.show()
```

# Funzione pendenza $m(x)$

```python
def m(x):
    dx=1e-5
    return (f(x+dx)-f(x-dx))/(2*dx)
```

```python
# Calcolo la pendenza m(x) per tutti i valori di x
m_values = m(x)

# Traccio f(x) e m(x) sullo stesso grafico
plt.plot(x, y, label='f(x)')
plt.plot(x, m_values, color='red', label='m(x)')
plt.xlabel('x')
plt.ylabel('Valore')
plt.title('f(x) e m(x) per x in [-10, 10]')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# m(x)  la derivata numerica di f(x), che  un polinomio di terzo grado.
# Quindi m(x)  un polinomio di secondo grado: m(x) = 3x^2 + 2x - 5
# Troviamo i coefficienti tramite np.polyfit

coeff_m = np.polyfit(x, m_values, 2)
print("I coefficienti della parabola m(x) sono:", coeff_m)
```
